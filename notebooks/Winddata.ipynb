```python
import zipfile
import os

zip_file_path = "/content/Hackathon_data (1).zip"

# Directory to extract the contents of the zip file
extract_dir = "content/"

# Check if the extraction directory exists, if not, create it
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# List the extracted files
extracted_files = os.listdir(extract_dir)
print("Extracted files:")
for file in extracted_files:
    print(file)
```

```python
import pandas as pd

# List of filenames for air temperature data
air_temp_files = [
    '/content/content/Hackathon_data/wind_power_data/air_temperature_2019.xlsx',
    '/content/content/Hackathon_data/wind_power_data/air_temperature_2020.xlsx',
    '/content/content/Hackathon_data/wind_power_data/air_temperature_2021.xlsx',
    '/content/content/Hackathon_data/wind_power_data/air_temperature_2022.xlsx',
    '/content/content/Hackathon_data/wind_power_data/air_temperature_2023.xlsx'
]

# Define column names for the air temperature data
column_names = ['DateTime', 'Air temperature | (°C)']

# Create an empty list to store individual DataFrames
dfs = []

# Read each file, specify column names, and append its DataFrame to the list
for file in air_temp_files:
    df = pd.read_excel(file, names=column_names, header=0)
    dfs.append(df)

# Concatenate all DataFrames in the list along rows
combined_air_temp_df = pd.concat(dfs, ignore_index=True)

# Print the combined DataFrame
print(combined_air_temp_df)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define SARIMA model parameters
order = (1, 1, 1)  # (p, d, q)
seasonal_order = (1, 1, 1, 12)  # (P, D, Q, S)

# Fit SARIMA model to the data
sarima_model = SARIMAX(df['Power generated by system | (MW)'], order=order, seasonal_order=seasonal_order)
sarima_result = sarima_model.fit()

# Print summary of the SARIMA model
print(sarima_result.summary())
```
