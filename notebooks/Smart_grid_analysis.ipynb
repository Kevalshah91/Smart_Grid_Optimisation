```python
df_pred['DateTime'] = pd.to_datetime(df_pred['DateTime'])

# Sort the dataframe by DateTime
df_pred.sort_values(by='DateTime', inplace=True)

# Calculate the lengths for each split based on percentages
total_length = len(df_pred)
length_1 = int(total_length * 0.20)
length_2 = int(total_length * 0.45)
length_3 = total_length - length_1 - length_2

# Extract corresponding dates for the first split
dates_gen_1 = df_pred['DateTime'].iloc[:length_1].reset_index(drop=True)

# Create a DataFrame with the date column as the first column
power_gen_df = pd.DataFrame({'date': dates_gen_1})

# Add the split 'Power_predicted_values' columns
power_gen_df['power_gen_1'] = df_pred['Power_predicted_values'].iloc[:length_1].reset_index(drop=True)
power_gen_df['power_gen_2'] = df_pred['Power_predicted_values'].iloc[length_1:length_1+length_2].reset_index(drop=True)
power_gen_df['power_gen_3'] = df_pred['Power_predicted_values'].iloc[-length_3:].reset_index(drop=True)

# Display the new dataframe
print(power_gen_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = grid_data.drop(columns=['stability','Power_Training_Predicted_Values','DateTime'])
y = grid_data['stability']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf_classifier = RandomForestClassifier()

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
test_grid.columns
```

```python
X_grid_test = test_grid.drop(columns=['stability','DateTime'])
y_grid_test = test_grid['stability']

y_grid_test_pred = rf_classifier.predict(X_grid_test)

accuracy_grid_test = accuracy_score(y_grid_test, y_grid_test_pred)
print("Accuracy on Test Set (grid_test):", accuracy_grid_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier

logistic_reg = LogisticRegression()
svm_classifier = SVC()
gradient_boosting = GradientBoostingClassifier()

logistic_reg.fit(X_train, y_train)
svm_classifier.fit(X_train, y_train)
gradient_boosting.fit(X_train, y_train)

y_pred_logistic_reg = logistic_reg.predict(X_grid_test)
y_pred_svm = svm_classifier.predict(X_grid_test)
y_pred_gradient_boosting = gradient_boosting.predict(X_grid_test)

accuracy_logistic_reg = accuracy_score(y_grid_test, y_pred_logistic_reg)
accuracy_svm = accuracy_score(y_grid_test, y_pred_svm)
accuracy_gradient_boosting = accuracy_score(y_grid_test, y_pred_gradient_boosting)

print("Accuracy of Logistic Regression:", accuracy_logistic_reg)
print("Accuracy of SVM:", accuracy_svm)
print("Accuracy of Gradient Boosting:", accuracy_gradient_boosting)
```
